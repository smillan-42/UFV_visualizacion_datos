```python
import plotly_express as px
```

```python
gapminder = px.data.gapminder()
gapminder2007 = gapminder.query("year == 2007") # filtramos solo los datos del 2007
```

Mostramos set de datos:

*   gdpPercap = producto interior bruto (cantidad riqueza del país dividido entre la población)



```python
gapminder.head()
```

```python
gapminder2007.head()
```

Gráfico de dispersión

```python
# los espacios dentro de la llamada a la funcion no se añade espacios
# gpaminder2007 se lo asigna directamente al 1º parametro
# la x si que la estoy llamando concretamente, por lo que esta daria igual donde se coloca en la llamada)

# los parametros que no se definan se les pasa lo que este definido por defecto
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp")
fig.show()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color="continent", hover_name="country")
fig.show()
```

Añado títulos ejes de la gráfica (diccionario)

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color="continent", hover_name="country", labels={"lifeExp":"Life expectation", "gdpPercap":"GDP per capita (USD)"})
fig.show()
```

Añadimos otra dimensión ---> tamaño población

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color="continent", hover_name="country", labels={"lifeExp":"Life expectation", "gdpPercap":"GDP per capita (USD)"}, size="pop", size_max=60)
# pop es la poblacion
# tamaño burbujas es proporcional a la poblacion
fig.show()
```

facet col = permite dividir el grafico pasando columna del dataframe

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color="continent", hover_name="country",
                 labels={"lifeExp":"Life expectation", "gdpPercap":"GDP per capita (USD)"}, size="pop", size_max=60, facet_col="continent")
# esto es otra forma de añadir otra dimension si hay muchas categorias / continentes
fig.show()
```
