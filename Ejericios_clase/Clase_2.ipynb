Transformo la columna FORECAST para que solo sean enteros y poder tener los meses directamente

```python
import re

def extract_number(text):
    match = re.search(r'\d+', text)
    if match:
        return int(match.group(0)) - 1
    else:
        return 12  # or any other default value if no number is found

df_forecast['FORECAST'] = df_forecast['FORECAST'].apply(extract_number)
```
